# Step4: Visualization

## Uni varient analysis

### Histograms
plotted for all numerical columns (like loan amount, interest rate, DTI, annual income, etc.)

```python
columns_count = len(num_col)
rows_count = (columns_count + 2) // 3 

fig, axes = plt.subplots(rows_count, 3, figsize=(15, 5 * rows_count))
axes = axes.flatten()
for idx, column in enumerate(num_col):
    df[column].plot(kind='hist', ax=axes[idx], bins=10, alpha=0.8, color='green', edgecolor='black')
    axes[idx].set_title(f'{column} Histogram')
    axes[idx].set_xlabel(column)
    axes[idx].set_ylabel('Frequency')

for unused_idx in range(idx + 1, len(axes)):
    fig.delaxes(axes[unused_idx])
plt.tight_layout()
plt.show()
```

### Insight:
- Loan amount is right-skewed  most loans are between $5,000 - $15,000.

- Interest rate is concentrated between 10% and 15%.

- Annual income is heavily right-skewed  most borrowers earn between $40,000 to $80,000.

- Some financial ratios like DTI are clustered around 10%-20%

### Box Plots
plots for detecting outliers in numerical columns

```python
columns_count = len(num_col)
rows_count = (columns_count + 2) // 3 

fig, axes = plt.subplots(rows_count, 3, figsize=(15, 5 * rows_count))
axes = axes.flatten()
for idx, column in enumerate(num_col):
    df[column].plot(kind='box', ax=axes[idx],color='red')
    axes[idx].set_title(f'{column} box plot')
    axes[idx].set_xlabel(column)
    axes[idx].set_ylabel('Frequency')

for unused_idx in range(idx + 1, len(axes)):
    fig.delaxes(axes[unused_idx])
plt.tight_layout()
plt.show()
```

### Insight:

- There are significant outliers in loan amount, annual income, and DTI.

- Annual income especially shows extreme outliers (some borrowers have very high income compared to the majority).

### KDE Plots 
plots for numerical columns like loan amount, interest rate, annual income, DTI, etc.

```python


columns_count = len(num_col)
rows = (columns_count // 3) + 1  

fig, axes = plt.subplots(rows, 3, figsize=(15, 5 * rows))  
axes = axes.flatten()  
for i, col in enumerate(num_col):
    sns.kdeplot(data=df, x=col, ax=axes[i], fill=True, color='blue', alpha=0.5)
    axes[i].set_title(f'KDE Plot of {col}')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Density')

for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])  

plt.tight_layout()  
plt.show()
```

### Insights:

- Loan Amount KDE: Smooth curve shows the most common loan amount is around $10,000, with very few large loans.

- Interest Rate KDE: Majority of loans cluster around 12%-15% interest rate. There's a long tail towards higher interest rates.

- Annual Income KDE: Strong right skew  Most borrowers earn around $50,000$80,000 per year.

- DTI KDE: Peaks around 15%, indicating most borrowers' debt is about 15% of their income.

## Performing bivariate analysis to explore relationships between loan status and numerical features


```python
plt.figure(figsize=(15, 10))
bcol = ['loan_amnt', 'annual_inc', 'dti', 'installment']

for i, feature in enumerate(bcol):
    plt.subplot(2, 2, i + 1)
    sns.boxplot(x='loan_status', y=feature, data=df)
    plt.title(f'Loan Status vs {feature}')
plt.tight_layout()
plt.show()  
```

### Insights from the Boxplots:
- Loan Amount vs. Loan Status: Borrowers who are fully paid tend to have a wider range of loan amounts compared to those who are charged off. The median loan amount for charged-off loans is lower.
- Annual Income vs. Loan Status: There is a noticeable difference in annual income between the two groups. Borrowers who fully paid their loans generally have higher annual incomes compared to those who were charged off.
- Debt-to-Income Ratio vs. Loan Status: The debt-to-income ratio shows that charged-off borrowers tend to have higher DTI ratios, indicating that they may be over-leveraged compared to those who fully paid.
- Installment vs. Loan Status: The installment amounts are generally higher for fully paid loans, suggesting that borrowers who can afford higher installments are more likely to pay off their loans.

## Performing bivariate analysis for categorical features against loan status

```python
plt.figure(figsize=(15, 10))
c_col = ['purpose', 'home_ownership', 'emp_length']

for i, feature in enumerate(c_col):
    plt.subplot(2, 2, i + 1)
    sns.countplot(x=feature, hue='loan_status', data=df, palette='Set2')
    plt.title(f'Loan Status by {feature}')
plt.tight_layout()
plt.show()  
```

### Insights from the Count Plots:
- Loan Purpose vs. Loan Status: The majority of loans for purposes such as credit card and debt consolidation have a higher proportion of fully paid statuses. In contrast, loans for small business purposes show a higher proportion of charged-off statuses.
- Home Ownership vs. Loan Status: Renters appear to have a higher proportion of charged-off loans compared to homeowners. This suggests that home ownership may be associated with better loan repayment behavior.
- Employment Length vs. Loan Status: Borrowers with longer employment lengths (10+ years) tend to have a higher proportion of fully paid loans, while those with less than 1 year of employment show a higher proportion of charged-off loans.

# Step5: Conclusion

- Most borrowers are of moderate risk profiles (Grades B and C), with stable income and long employment histories. 
- Debt Consolidation is the most common reason for loans. 
- Fully Paid loans dominate, but a non-negligible portion of Charged Off loans indicates scope for better risk modeling.
