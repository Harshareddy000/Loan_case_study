# Step4: Visualization

## Uni varient analysis

### Histograms
plotted for all numerical columns (like loan amount, interest rate, DTI, annual income, etc.)

```python
columns_count = len(num_col)
rows_count = (columns_count + 2) // 3 

fig, axes = plt.subplots(rows_count, 3, figsize=(15, 5 * rows_count))
axes = axes.flatten()
for idx, column in enumerate(num_col):
    df[column].plot(kind='hist', ax=axes[idx], bins=10, alpha=0.8, color='green', edgecolor='black')
    axes[idx].set_title(f'{column} Histogram')
    axes[idx].set_xlabel(column)
    axes[idx].set_ylabel('Frequency')

for unused_idx in range(idx + 1, len(axes)):
    fig.delaxes(axes[unused_idx])
plt.tight_layout()
plt.show()
```

### Insight:
- Loan Amounts and Funding: Most loans and funded amounts peak around 10,000, with a right-skew indicating fewer large loans extending up to 35,000.

- Interest Rates and Installments: Interest rates cluster between 10-15%, with a peak around 12%. Installments are concentrated between 200-400, showing manageable monthly payments for most borrowers.

- Annual Income and Debt Ratios: Most borrowers have incomes below 100,000, with some outliers up to 6,000,000. Debt-to-income ratios (DTI) peak around 15, reflecting moderate debt levels among borrowers.

- Credit Activity: Most borrowers have 0-2 recent credit inquiries (inq_last_6mths), 5-10 open accounts, and low revolving utilization rates (20-40%), indicating stable credit behavior.

- Risk Indicators: Public records and bankruptcies are heavily concentrated at zero, highlighting minimal financial distress among borrowers.

- Credit History: Total credit accounts peak around 20, reflecting moderate credit histories, with some borrowers having extensive histories.

### KDE Plots

## Bivariate Analysis

## Performing bivariate analysis to explore relationships between loan status and numerical features


```python
plt.figure(figsize=(15, 10))
bcol = ['loan_amnt', 'annual_inc', 'dti', 'installment']

for i, feature in enumerate(bcol):
    plt.subplot(2, 2, i + 1)
    sns.boxplot(x='loan_status', y=feature, data=df)
    plt.title(f'Loan Status vs {feature}')
plt.tight_layout()
plt.show()  
```

### Insights from the Boxplots:
- Loan Amount vs. Loan Status: Charged-off loans have a higher median loan amount compared to fully paid or current loans, suggesting larger loans may carry a higher risk of default.

- Annual Income vs. Loan Status: Median annual income remains similar across loan statuses, but outliers with high incomes in the "Fully Paid" category could indicate quick repayments by high-income borrowers.

- DTI vs. Loan Status: Median DTI is consistent across statuses, but charged-off loans exhibit a wider spread, suggesting that extreme DTI values might increase default risk.

- Installment vs. Loan Status: Charged-off loans show slightly higher median installment amounts, which aligns with their larger loan sizes.

- Outliers: Extreme values in all plots highlight cases that may require deeper examination to understand their impact on loan outcomes.

- Further Considerations: The box plots provide valuable insights but cannot establish causation. Additional analysis and factors are needed for a complete risk assessment.
