## Performing bivariate analysis to explore relationships between loan status and numerical features


```python
plt.figure(figsize=(15, 10))
bcol = ['loan_amnt', 'annual_inc', 'dti', 'installment']

for i, feature in enumerate(bcol):
    plt.subplot(2, 2, i + 1)
    sns.boxplot(x='loan_status', y=feature, data=df)
    plt.title(f'Loan Status vs {feature}')
plt.tight_layout()
plt.show()  
```

### Insights from the Boxplots:
Loan Amount vs. Loan Status: Borrowers who are fully paid tend to have a wider range of loan amounts compared to those who are charged off. The median loan amount for charged-off loans is lower.
Annual Income vs. Loan Status: There is a noticeable difference in annual income between the two groups. Borrowers who fully paid their loans generally have higher annual incomes compared to those who were charged off.
Debt-to-Income Ratio vs. Loan Status: The debt-to-income ratio shows that charged-off borrowers tend to have higher DTI ratios, indicating that they may be over-leveraged compared to those who fully paid.
Installment vs. Loan Status: The installment amounts are generally higher for fully paid loans, suggesting that borrowers who can afford higher installments are more likely to pay off their loans.

## Performing bivariate analysis for categorical features against loan status

```python
plt.figure(figsize=(15, 10))
c_col = ['purpose', 'home_ownership', 'emp_length']

for i, feature in enumerate(c_col):
    plt.subplot(2, 2, i + 1)
    sns.countplot(x=feature, hue='loan_status', data=df, palette='Set2')
    plt.title(f'Loan Status by {feature}')
plt.tight_layout()
plt.show()  
```

### Insights from the Count Plots:
Loan Purpose vs. Loan Status: The majority of loans for purposes such as credit card and debt consolidation have a higher proportion of fully paid statuses. In contrast, loans for small business purposes show a higher proportion of charged-off statuses.
Home Ownership vs. Loan Status: Renters appear to have a higher proportion of charged-off loans compared to homeowners. This suggests that home ownership may be associated with better loan repayment behavior.
Employment Length vs. Loan Status: Borrowers with longer employment lengths (10+ years) tend to have a higher proportion of fully paid loans, while those with less than 1 year of employment show a higher proportion of charged-off loans.


```python
# Performing multivariate analysis using scatterplots to explore relationships between multiple numerical features
# Focusing on the relationship between loan amount, annual income, and loan status

# Setting the figure size
plt.figure(figsize=(10, 6))

# Scatterplot of loan amount vs. annual income colored by loan status
sns.scatterplot(x='loan_amnt', y='annual_inc', hue='loan_status', data=df, alpha=0.6)
plt.title('Loan Amount vs. Annual Income by Loan Status')
plt.xlabel('Loan Amount')
plt.ylabel('Annual Income')
plt.legend(title='Loan Status')
plt.show()  
```

### Insights from the Scatterplot:
The scatterplot shows that borrowers with higher annual incomes tend to take larger loan amounts.
There is a clear distinction between the loan statuses: borrowers who fully paid their loans (in green) are more concentrated in the higher income and loan amount regions, while those who defaulted (in red) are more spread out, particularly in the lower income and loan amount areas.
This suggests that higher income may be associated with a lower likelihood of defaulting on loans.
